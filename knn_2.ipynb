```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
from sklearn import datasets
dataset_iris=datasets.load_iris()
x=dataset_iris['data'].copy()
y=dataset_iris['target'].copy()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=666)
```

```python
train_x
```

```python
from sklearn.preprocessing import StandardScaler
knn_StandardScaler=StandardScaler()
knn_StandardScaler.fit(train_x)
train_x=knn_StandardScaler.transform(train_x)
test_x=knn_StandardScaler.transform(test_x)
knn_StandardScaler.mean_
knn_StandardScaler.scale_
```

## 网格搜索

```python
param_grid=[
    {'n_neighbors':[k for k in range(1,11)],
    'weights':['uniform']},
    {'n_neighbors':[k for k in range(1,11)],
    'weights':['distance'],
    'p':[p for p in range(1,11)]}
]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_cls=KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
knn_GridSearchCV= GridSearchCV(knn_cls,param_grid,verbose=2)
knn_GridSearchCV.fit(train_x,train_y)
```

```python
knn_GridSearchCV.best_params_
```

```python
knn_cls=KNeighborsClassifier(n_neighbors=6,weights='distance',p=3)
knn_cls.fit(train_x,train_y)
```

```python
knn_cls.score(test_x,test_y)
```

```python
x_predict=np.array([4.4,  2.0 ,  3.4,  4.2])
x_predict_standard=knn_StandardScaler.transform(x_predict)
x_predict=x_predict.reshape(1,-1)
knn_cls.predict(x_predict)
```
